```python

import pandas as pd

df = pd.read_csv('C:/Users/sachi/Downloads/lego_sets.csv')
theme = pd.read_csv('C:/Users/sachi/Downloads/parent_themes.csv')

theme.head()
```

```python
merged = df.merge(theme, left_on ='parent_theme', right_on = 'name')
merged.head()
```

```python
merged.drop(columns='name_y', inplace = True)
merged.head()
```

```python
licensed = merged[merged['is_licensed']]
licensed.head()
```

```python
# Question-1:-What percentage of all licensed sets ever released were star Wars themed?
star_wars = licensed[licensed['parent_theme']=='Star Wars']
star_wars.head()
```

```python
#checking how many null values present in set_num column
merged[merged['set_num'].isnull()].shape
```

```python
# storing the percentage in the specified {'the_force'} variable
licensed = merged[merged['is_licensed']]
licensed = licensed.dropna(subset=['set_num'])
licensed.head()

the_force = int(star_wars.shape[0]/licensed.shape[0]*100)

print(the_force)
```
